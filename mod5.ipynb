```python
# Загрузите файл wage_data.csv в DataFrame. 
# Постройте линейный график изменения минимальной зарплаты (признак Salary), усредненной по всем штатам

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('wage-data.csv')
df
```

```python
series = df.groupby('Year')['Salary'].agg(np.mean)
x_range_list = list(range(series.index.min(), series.index.max(), 5)) + [series.index.max()]
plt.plot(series)
plt.title('\nИзменения минимальной зарплаты, усредненной по всем штатам\n')
plt.xlabel('\nГод\n')
plt.ylabel('\nЗаработная плата $ / час\n')
plt.xticks(x_range_list, rotation=45)
```

```python
plt.scatter(df['Year'], df['Salary'])
plt.title('\nИзменения минимальной зарплаты, усредненной по всем штатам\n')
plt.xlabel('\nГод\n')
plt.ylabel('\nЗаработная плата $ / час\n')
plt.xticks(x_range_list, rotation=45)
```
